```python
# Basically, if you see allwcs above, its list of all wcs
# Here, it is extracting all the wilson coeffs which doesnt have an '_' in their name
# In fact, all wcs that have an index (ijkl etc), they are written after '_'Applications
# Example: uu_1234 (4 numbers, 4 fermions), uphi_22 (2 numbers, 2 fermions)
wckeys0f=list(dict.fromkeys(v for v in allwcs if "_" not in v))
print(wckeys0f)
```

```python
# First it splits the wc at '_'. For example: uphi_ij (i,j are numbers).
#On splitting, the first argument is uphi, which goes into naming
# but since it wants only 2fermion ones here, it checks if the length of second agrument after split (i.e. ij)
# has length 2. If yes, it goes here. Otherwise, we make other list where 4fermion wcs are listed 
wckeys2f=list(
            dict.fromkeys(
                v.split("_")[0]
                for v in allwcs
                if "_" in v and len(v.split("_")[1]) == 2
            )
        )
print(wckeys2f)
```

```python
# same as above, but for 4 fermions
wckeys4f=list(
            dict.fromkeys(
                v.split("_")[0]
                for v in allwcs
                if "_" in v and len(v.split("_")[1]) == 4
            )
        )
print(wckeys4f)
```

```python
#now we have all wc names, without the numbers (ex. we only have uu, not uu_1111, uu_2222 etc.)
wckeys=wckeys0f+wckeys2f+wckeys4f
```

```python
#the output here makes it apparent what this section of the code is doing - we assign to the key some values
#these values are basically all the fermion generation combinations : 1111, 1122 etc.
indexdict = {k: [] for k in wckeys}
for v in allwcs:
            v_split = v.split("_")
            if len(v_split) == 2:
                indexdict[v_split[0]].append(v_split[1])
print(indexdict)
            
```

```python
#this function gives the shape of all keys - 1, (3,3) or (3,3,3,3) 
#tuple, is a single variable that is used to store multiple dictionary 
# In python, the other three data types that do is are dictionary, list and set (fourth is tuple)
#in tuple, order is unchangeable, and are written with round bracket. THEY ALLOW DUPLICATES UNLIKE DICTIONARY 
#example: mytuple = ("item1","item2","item3") 
#see the cells below to understand the functioning in detail
import numpy as np # need to import this package to use the math funcitons
wckeysshape = {
            k: tuple(
                np.max(
                    [[int(i) for i in index_list] for index_list in v], axis=0
                )
            )
            if v
            else 1
            for k, v in indexdict.items()
        }
print(wckeysshape)
```
